```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read data from the text file
with open('vector.txt', 'r') as file:
    data = file.read().split()  # Read the data and split by spaces or commas

# v = data[int(len(data)/2)]
data_x = data[1:int(len(data)/2)]
data_v = data[int(len(data)/2)+1::]

# Convert the data to numerical values (if necessary)
data_x = np.array([float(val) for val in data_x])  # Convert strings to floats
data_v = np.array([float(val) for val in data_v])  # Convert strings to floats

data_x1 = data_x + 1
data_x2 = data_x + 2
data_x3 = data_x - 1
data_x4 = data_x - 2
data_x5 = data_x + 3
data_x6 = data_x - 3
data_x7 = data_x + 4
data_x8 = data_x - 4
data_x9 = data_x + 5
data_x10 = data_x - 5
data_x11 = data_x + 6
data_x12 = data_x - 6
```

```python
# Create a simple plot
plt.figure(figsize=(20, 12))
plt.plot(data_x, data_v)
plt.plot(data_x1, data_v)
plt.plot(data_x2, data_v)
plt.plot(data_x3, data_v)
plt.plot(data_x4, data_v)
plt.plot(data_x5, data_v)
plt.plot(data_x6, data_v)
plt.plot(data_x7, data_v)
plt.plot(data_x8, data_v)
plt.plot(data_x9, data_v)
plt.plot(data_x10, data_v)
plt.plot(data_x11, data_v)
plt.plot(data_x12, data_v)
plt.xlim(0,2) 
plt.ylim(-1,1) 
plt.xlabel('x')
plt.ylabel('v')
plt.title('Data Plot')
plt.show()
```
